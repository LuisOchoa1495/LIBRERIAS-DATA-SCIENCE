### 1. IMPORTAR LIBRERIAS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 2. LEER HTML

```python
tables=pd.read_html('https://as.com/especiales/futbol/copa_america/comunes/palmares/')
print(f'Número de tablas: {len(tables)}')
```

### 3. MOSTRAR TODAS LAS TABLAS

```python
for i in range(len(tables)):
    print(tables[i])
```

### 4. SELECCIONAR DATA

```python
datos=(tables[0])
print(datos)
```

### 5. CREAR DATAFRAME

```python
dataframe=pd.DataFrame(datos)
dataframe
```

### 6. IMPORTAR CSV

```python
if(True):
    dataframe.to_csv('.\copa_america_ediciones.csv')
    print('SE IMPORTO CORRECTAMENTE A FORMATO CSV')
```

### 7. CONTAR VALORES

```python
df=dataframe['Campeón'].value_counts()
df=df.reset_index()
df.columns=['Pais', 'Campeonatos']
df
```

### 8. GRÁFICA CIRCULAR

```python
plt.pie(df['Campeonatos'],labels=df['Pais'],autopct='%0.1f%%',pctdistance=0.7)
plt.title("PORCENTAJE DE CAMPEONATOS - COPA AMERICA")
plt.show()
```
