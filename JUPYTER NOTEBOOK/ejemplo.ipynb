## IMPORTAR LIBRERIAS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## LEER HTML

```python
#leer tablas 
tables=pd.read_html('https://www.espn.com.pe/futbol/copa-america/nota/_/id/243961/copa-america-historia-campeon-subcampeon-ganadores-titulos')
print(f'Numero de tablas: {len(tables)}')
```

## MOSTRAR TODAS LAS TABLAS

```python
#imprimir tablas
for i in range(len(tables)):
    print(tables[i])
```

## SELECCIONAR DATA 

```python
#datos
datos=(tables[0])
print(datos)
```

## CREAR DATAFRAME

```python
dataframe=pd.DataFrame(datos)
dataframe
```

## IMPORTAR CSV

```python
if(True):
    dataframe.to_csv('.\copa_america_campeones.csv')
    print('Se importo correctamente a formato csv')
```

## CONTAR VALORES

```python
df=dataframe['CAMPEÓN'].value_counts()
df=df.reset_index()
df.columns=['Pais', 'Total']
df
```

## REALIZAR GRÁFICA

```python
borde={'linewidth':0.1 , 'edgecolor':"black"}
plt.pie(df['Total'],labels=df['Pais'],explode=(0.1,0,0,0,0,0,0,0),autopct='%0.1f%%',pctdistance=0.7)
plt.title("PORCENTAJE DE TITULOS - COPA AMERICA")
plt.show()
```
